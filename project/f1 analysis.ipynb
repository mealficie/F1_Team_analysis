# Extracting Data from Wikipedia and Cleaning it

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib widget

plt.style.use('seaborn-colorblind')
```

```python
#Cleaning the red bull dataset 
from matplotlib.axis import XAxis
from matplotlib.pyplot import xlim


red_bull_df_1=pd.read_html("Red Bull Racing Grand Prix results - Wikipedia.html")[1]
red_bull_df_2=pd.read_html("Red Bull Racing Grand Prix results - Wikipedia.html")[2]

red_bull_df=red_bull_df_1.append(red_bull_df_2,ignore_index=True)

red_bull=pd.DataFrame()
red_bull['Year']=red_bull_df['Year'].astype(str)
red_bull['Points']=red_bull_df['Points']
red_bull['WCC']=red_bull_df['WCC'].str.strip()
red_bull.drop_duplicates(subset=None,keep='last',inplace=True)
red_bull['WCC']=red_bull['WCC'].str.replace("[A-Za-z]","")
red_bull.drop([40,41],0,inplace=True)

#red_bull.set_index("Year",inplace=True)
red_bull["WCC"]=red_bull["WCC"].astype(int)
red_bull["Points"]=red_bull["Points"].astype(float)

red_bull.head(20)
```

```python
#cleaning the Mercedes Dataset
mercedes_df_1=pd.read_html("Mercedes-Benz Grand Prix results - Wikipedia.html")[2]
mercedes_df_2=pd.read_html("Mercedes-Benz Grand Prix results - Wikipedia.html")[3]
mercedes_df=mercedes_df_1.append(mercedes_df_2,ignore_index=True)

mercedes_df.head()
mercedes=pd.DataFrame()
mercedes['Year']=mercedes_df['Year'].astype(str)
mercedes['Points']=mercedes_df['Points']
mercedes['WCC']=mercedes_df['WCC'].str.strip()
mercedes.drop_duplicates(subset=None,keep='last',inplace=True)
mercedes['WCC']=mercedes['WCC'].str.replace("[A-Za-z]","")

mercedes.drop([40,41],0,inplace=True)

#mercedes.set_index("Year",inplace=True)
mercedes["WCC"]=mercedes["WCC"].astype(int)
mercedes["Points"]=mercedes["Points"].astype(float)

mercedes.head(20)
```

```python
#Cleaning the ferrari dataset
ferrari_df_1=pd.read_html("Ferrari Grand Prix results - Wikipedia.html")[6]
ferrari_df_2=pd.read_html("Ferrari Grand Prix results - Wikipedia.html")[7]
ferrari_df=ferrari_df_1.append(ferrari_df_2,ignore_index=True)

#ferrari_df.head()

ferrari=pd.DataFrame()
ferrari['Year']=ferrari_df['Year'].astype(str)
ferrari['Points']=ferrari_df['Points']
ferrari['WCC']=ferrari_df['WCC'].str.strip()
ferrari.drop_duplicates(subset=None,keep='last',inplace=True)
ferrari['WCC']=ferrari['WCC'].str.replace("[A-Za-z]","")
ferrari.drop([39,40],0,inplace=True)

#ferrari.set_index("Year",inplace=True)

ferrari["WCC"]=ferrari["WCC"].astype(int)
ferrari["Points"]=ferrari["Points"].astype(float)


ferrari.head(20)
```

```python
#Cleaning the ferrari dataset
mclaren_df_1=pd.read_html("McLaren Grand Prix results - Wikipedia.html")[5]
mclaren_df_2=pd.read_html("McLaren Grand Prix results - Wikipedia.html")[6]
mclaren_df=mclaren_df_1.append(mclaren_df_2,ignore_index=True)

mclaren_df.head()
mclaren=pd.DataFrame()
mclaren['Year']=mclaren_df['Year'].astype(str)
mclaren['Points']=mclaren_df['Points']
mclaren['WCC']=mclaren_df['WCC'].str.strip()
mclaren.drop_duplicates(subset=None,keep='last',inplace=True)
mclaren['WCC']=mclaren['WCC'].str.replace("[A-Za-z]","")

mclaren.drop([33,42,43],0,inplace=True)

#mclaren.set_index("Year",inplace=True)
mclaren["WCC"]=mclaren["WCC"].astype(int)
mclaren["Points"]=mclaren["Points"].astype(float)


mclaren.head(20)
```

# Visulizing Line Plot for different teams with passing year

```python
from turtle import color

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
sns.pointplot(red_bull['Year'], red_bull['Points'],ax=ax1,ci=None )
ax1.title.set_text('For Red Bull')
ax1.set_xticks(ax1.get_xticks()[::2])
ax1.tick_params(top=False, left=False, right=False, labelleft='off', labelbottom='on')
avg_red_bull=red_bull['Points'].sum()/len(red_bull['Year'])

#Beautifying the red bull subplot
ax1.axhline(y=avg_red_bull,color='gray',linewidth=1,linestyle='--')
ax1.fill_between(range(len(red_bull['Year'])),red_bull['Points'],avg_red_bull, where=red_bull['Points']>=avg_red_bull, color='purple',interpolate=True,alpha=0.3)
ax1.fill_between(range(len(red_bull['Year'])),red_bull['Points'],avg_red_bull, where=red_bull['Points']<=avg_red_bull, color='yellow',interpolate=True,alpha=0.3)
ax1.spines.right.set_visible(False)
ax1.spines.top.set_visible(False)

sns.pointplot(mercedes['Year'], mercedes['Points'], color='silver',ax=ax2,ci=None)
ax2.title.set_text('For Mercedes')
ax2.set_xticks(ax2.get_xticks()[::2])
ax2.tick_params(top=False, left=False, right=False, labelleft='off', labelbottom='on')
avg_mercedes=mercedes['Points'].sum()/len(mercedes['Year'])
#Beautifying the mercedes subplot
ax2.axhline(y=avg_mercedes,color='gray',linewidth=1,linestyle='--')
ax2.fill_between(range(len(mercedes['Year'])),mercedes['Points'],avg_mercedes, where=mercedes['Points']>=avg_mercedes, color='purple',interpolate=True,alpha=0.3)
ax2.fill_between(range(len(mercedes['Year'])),mercedes['Points'],avg_mercedes, where=mercedes['Points']<=avg_mercedes, color='yellow',interpolate=True,alpha=0.3)
ax2.spines.right.set_visible(False)
ax2.spines.top.set_visible(False)


sns.pointplot(ferrari['Year'], ferrari['Points'], color='red',ax=ax3,ci=None)
ax3.title.set_text('For Ferrari')
ax3.set_xticks(ax3.get_xticks()[::2])
ax3.tick_params(top=False, left=False, right=False, labelleft='off', labelbottom='on')
avg_ferrari=ferrari['Points'].sum()/len(ferrari['Year'])
ax3.axhline(y=avg_ferrari,color='gray',linewidth=1,linestyle='--')

#Beautifying the ferrari subplot
ax3.fill_between(range(len(ferrari['Year'])),ferrari['Points'],avg_ferrari, where=ferrari['Points']>=avg_ferrari, color='purple',interpolate=True,alpha=0.3)
ax3.fill_between(range(len(ferrari['Year'])),ferrari['Points'],avg_ferrari, where=ferrari['Points']<=avg_ferrari, color='yellow',interpolate=True,alpha=0.3)
ax3.spines.right.set_visible(False)
ax3.spines.top.set_visible(False)


sns.pointplot(mclaren['Year'], mclaren['Points'], color='orange',ax=ax4,ci=None)
ax4.title.set_text('For Mclaren')
ax4.set_xticks(ax4.get_xticks()[::2])
ax4.tick_params(top=False, left=False, right=False, labelleft='off', labelbottom='on')
avg_mclaren=mclaren['Points'].sum()/len(mclaren['Year'])

#Beautifying the mclaren subplot
ax4.axhline(y=avg_mclaren,color='gray',linewidth=1,linestyle='--')
ax4.fill_between(range(len(mclaren['Year'])),mclaren['Points'],avg_mclaren, where=mclaren['Points']>=avg_mclaren, color='purple',interpolate=True,alpha=0.3)
ax4.fill_between(range(len(mclaren['Year'])),mclaren['Points'],avg_mclaren, where=mclaren['Points']<=avg_mclaren, color='yellow',interpolate=True,alpha=0.3)
ax4.spines.right.set_visible(False)
ax4.spines.top.set_visible(False)


fig.suptitle('Points Per Year for each team')


for ax in fig.get_axes():
    ax.label_outer()
```

# Visulizing Movement in Constructors Standings with each passing year

```python
#red_bull.set_index('Year')
#mercedes['Year']=pd.to_datetime(mercedes['Year'])
#ferrari['Year']=pd.to_datetime(ferrari['Year'])
#mclaren['Year']=pd.to_datetime(mclaren['Year'])

merged_data = pd.merge(red_bull.drop(['Points'],axis=1),mercedes.drop(['Points'],axis=1),on='Year')
merged_data=merged_data.merge(ferrari.drop(['Points'],axis=1),on='Year')
merged_data=merged_data.merge(mclaren.drop(['Points'],axis=1),on='Year')
merged_data.columns = ['Year','Red Bull', 'Mercedes', 'Ferrari', 'Mclaren']
merged_data

fig, ax = plt.subplots()

ax.plot(merged_data['Year'], merged_data['Red Bull'],label='Red Bull',color='Blue',marker ='*')
ax.plot(merged_data['Year'], merged_data['Mercedes'],label='Mercedes',color='Silver',marker ='*')
ax.plot(merged_data['Year'], merged_data['Ferrari'],label='Ferrari',color='Red',marker ='*')
ax.plot(merged_data['Year'], merged_data['Mclaren'],label='Mclaren',color='Orange',marker ='*')
ax.axis()
plt.tick_params(top=False, left=False, right=False, labelleft='off', labelbottom='on')
ax.spines.right.set_visible(False)
ax.spines.top.set_visible(False)

plt.gca().invert_yaxis()
ax.set_ylabel('Final Standing in Constructor Championship')
ax.set_xlabel('Year')
ax.set_title("Position in WCC vs Year")

ax.axvspan('2010','2014',color='gray',alpha=0.1,label='V8 2.4L Engine')
ax.axvspan('2014','2018',color='yellow',alpha=0.1,label='V6 1.6L Hybrid Engine')
ax.axvspan('2018','2021',color='purple',alpha=0.1,label='Halo Introduction')

plt.legend(loc='best',frameon=False, title='Teams and Rule Changes')

plt.show()

```
